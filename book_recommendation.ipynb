```python
import pandas as pd
books = pd.read_csv('/content/drive/MyDrive/archive2/books.csv')
book_tags = pd.read_csv('/content/drive/MyDrive/archive2/book_tags.csv')
tags = pd.read_csv('/content/drive/MyDrive/archive2/tags.csv')
```

```python
# combine all strings:
books_1 = books.drop(['id','best_book_id','work_id','books_count','isbn', 'isbn13', 'original_publication_year', 'original_title', 
                    'language_code', 'average_rating', 'ratings_count', 'work_ratings_count', 'work_text_reviews_count', 
                    'ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5'],axis=1)

books_1.head()
```

```python
import pandas as pd

# Gộp DataFrame bằng cột 'id'
merged_df = pd.merge(book_tags, tags, on='tag_id', how='inner')
merged_df.head()
```

```python
sorted_df = merged_df.sort_values('goodreads_book_id', ascending=True)

print(sorted_df)
```

```python
tag2 = sorted_df.drop(['tag_id','count'],axis=1)
tag2 = tag2.reset_index()
tag2 = tag2.drop('index',axis=1)
```

```python
print(tag2)
```

```python
tag3 = tag2.groupby('goodreads_book_id').agg({'tag_name': ' '.join}).reset_index()

print(tag3)
```

```python
df = pd.merge(books_1, tag3, left_on='book_id', right_on='goodreads_book_id', how='inner')

df.head()
```

```python
def clean_text(author):
  result = str(author).lower()
  result = result.replace(' ','')
  result = result.replace(',',' ')
  return(result)
```

```python
df['authors'] = df['authors'].apply(clean_text)
```

```python
df.head()
```

```python
df['data'] = df['authors'] + ' ' + df['tag_name']
```

```python
df = pd.DataFrame(data=df[['book_id', 'title', 'data', 'image_url', 'small_image_url']])
df.head()
```

```python
df_3 = pd.DataFrame(data=df[['book_id', 'title', 'image_url', 'small_image_url']])
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df['data'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df_1 = pd.DataFrame(similarities, columns=books['title'], index=df['title']).reset_index()
```

```python
df_1['book_id'] = df.book_id
```
